```python
#Renaming the columns
df = df.rename(columns={'Annual Income (k$)':'Annual Income','Spending Score (1-100)':'Spending Score'})
```

```python
df.head()
```

```python
#drop the CustomerID column
df.drop('CustomerID', axis=1, inplace=True)
df.head()
```

```python
#here replace Female as 0, and male as 1
df['Gender'].replace(['Female','Male'],[0,1],inplace=True)
```

```python
#check the gender column
df.head()
```

# EDA with Data Visualisation

```python
#Gender column
plt.figure(figsize=(12,6))
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
#Age column
plt.figure(figsize=(12,6))
sns.distplot(df['Age'])
plt.show()
```

```python
#Annual income column
plt.figure(figsize=(12,6))
sns.distplot(df['Annual Income'])
plt.show()
```

```python
sns.boxplot(data=df,orient='h',palette='rainbow_r')
```

```python
sns.pairplot(df,kind="reg", hue='Gender', palette='twilight')
plt.tight_layout()
```

The above pairplot graphic gives relation between different features while making a difference between genders.it can be seen that there is a relation between age,annual inacome,and spending score.

# Age vs Spending score 

```python
sns.lmplot(x="Age", y="Spending Score", data = df,hue="Gender",palette="rainbow_r")
```

Wow!!! Here we can clearly see that spending score is higher in age group of 20 to 40 ....as senior citizen (50 to 70) has less Spending score

# Annual Income vs Spending Score

```python
sns.lmplot(x="Annual Income", y="Spending Score", data = df,hue="Gender",palette="plasma")
```

Spending SCore is saturated at 40 to 60 for values of income between 40 to 70

# Average Income vs Age 

```python
sns.lmplot(x="Age", y="Annual Income", data = df,hue="Gender",palette="cool")
```
